```python
import pandas as pd
df = pd.read_csv('Bicycle_Thefts_Open_Data.csv')
df_bike_characteristics = df[['BIKE_MODEL', 'BIKE_TYPE', 'BIKE_SPEED', 'BIKE_COLOUR', 'BIKE_COST']].dropna()
bike_cost_stats = df_bike_characteristics['BIKE_COST'].describe()
bike_model_counts = df_bike_characteristics['BIKE_MODEL'].value_counts()
bike_type_counts = df_bike_characteristics['BIKE_TYPE'].value_counts()
bike_speed_counts = df_bike_characteristics['BIKE_SPEED'].value_counts()
bike_colour_counts = df_bike_characteristics['BIKE_COLOUR'].value_counts()
print("Basic Statistics for BIKE_COST:")
print(bike_cost_stats)
print("\nCounts for BIKE_MODEL:")
print(bike_model_counts)
print("\nCounts for BIKE_TYPE:")
print(bike_type_counts)
print("\nCounts for BIKE_SPEED:")
print(bike_speed_counts)
print("\nCounts for BIKE_COLOUR:")
print(bike_colour_counts)

```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv('Bicycle_Thefts_Open_Data.csv')
df_bike_characteristics = df[['BIKE_MODEL', 'BIKE_TYPE', 'BIKE_COLOUR']].dropna()
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))
sns.countplot(x='BIKE_MODEL', data=df_bike_characteristics, order=df_bike_characteristics['BIKE_MODEL'].value_counts().index[:5], ax=axes[0, 0])
axes[0, 0].set_title('Most Frequent BIKE_MODEL')
sns.countplot(x='BIKE_TYPE', data=df_bike_characteristics, order=df_bike_characteristics['BIKE_TYPE'].value_counts().index[:5], ax=axes[0, 1])
axes[0, 1].set_title('Most Frequent BIKE_TYPE')
sns.countplot(x='BIKE_COLOUR', data=df_bike_characteristics, order=df_bike_characteristics['BIKE_COLOUR'].value_counts().index[:5], ax=axes[1, 1])
axes[1, 1].set_title('Most Frequent BIKE_COLOUR')
plt.tight_layout()
plt.show()
```

From this graph, we can contast that the most common stolen bike profile is the next:
Unknown model,Type Mountain, and black
The type profile of a stolen bike is the next:
It has been stolen on July,a friday,outside, it's a moutain and black bike


We can conclude from these graphs and pies charts that the predictive factors are during the summer months,maybe because people let their bike at home, or tourist who are visiting don't think about the theft
We can also constat that the theft is more present in the summer, because during the winter, becaus of the rain and the cold,people are not more riding a bicycle
A prevention that we can give is to attach his bike outside with very good securities
And if you are going in holidays, give your bike to supervise to a friend or some family, and if you can't, protect it as much as you can, because those months are very critics for the theft,particularly in the night time, between 18:00 and 00:00
We can put forward the hypothesis that these thefts are targeting the tourists, or the locals who are on holidays and leave their bicycle without monitoring,especially during the evening hours
